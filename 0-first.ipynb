```python
import nltk
import ssl

try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context

nltk.download('punkt_tab')
```

```python
my_contd = ''' 
VIT Bhopal University, is a state private university of Madhya Pradesh situated in Kothri Kalan in Sehore district, Madhya Pradesh, India. It was established by the Vellore Institute of Technology. It is a world-renowned university and is often considered one of the absolute best in Madhya Pradesh.VIT Bhopal's worth depends on priorities: it offers strong placements (especially with high CGPA), good infrastructure, and industry-aligned curriculum, but some find its newness means lower peer quality/crowd than older campuses, plus potential issues with some faculty reliance on PPTs and location far from Bhopal city. It's a solid choice for those focused on career outcomes (good packages), faculty choice (FFCS), and research, but be prepared for rigorous academics and potentially less vibrant campus life compared to older VITs. 

'''
```

```python
#import sentence tokenizer ;  
# sent_tokenize --> sentence-tokenize ; word_tokenize --> word-tokenize

from nltk.tokenize import sent_tokenize

t_sent = sent_tokenize(my_contd)
t_sent
```

```python
#WORD-TOKENIZE
from nltk import word_tokenize
tknz_wrd = word_tokenize(my_contd)
tknz_wrd
```

```python
#word punctuation tokenize
from nltk.tokenize import wordpunct_tokenize
tknz_wrdpunc = wordpunct_tokenize(my_contd)
tknz_wrdpunc
```
