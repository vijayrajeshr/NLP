```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("If you want to be an excellent programmer \
, be consistent to practice daily on GFG.")

for token in doc:
    print(token, " | ",
          spacy.explain(token.pos_),
          " | ", token.lemma_)
```
