# Retrieval-Based Chatbot

This project builds a simple chatbot using **TF-IDF** and **cosine similarity**.

- The knowledge base is stored as a list of sentences.
- User input is converted into numerical vectors using `TfidfVectorizer()`.
- Cosine similarity is calculated between the user query and stored sentences.
- The sentence with the highest similarity score is returned as the response.

### Cosine Similarity Formula
It measures how close two text vectors are.
0 = Entirely Different
1 = Exactly The Same

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Simple Data (The Bot's Knowledge)
sentences = [
    "Perseus is identified as the high-level mole operating within the CIA.",
    "The biochemical weapon known as 'The Cradle' was intended to be launched at the Pentagon.",
    "The Pantheon group is a rogue paramilitary organization led by Jane Harrow.",
    "Jane Harrow was a high-ranking CIA official who secretly founded and commanded Pantheon.",
    "The shadow Black Ops team tasked with hunting Pantheon includes Frank Woods and Troy Marshall.",
    "Case is the primary field operative of the rogue Black Ops team whose identity remains mysterious.",
    "Svati and Felix provide technical and tactical support for the Black Ops team's unsanctioned missions.",
    "Adler was framed by Pantheon to lure the Black Ops team into a global conspiracy.",
    "The Safehouse, known as 'The Manor,' serves as the secret base of operations for Woods and Marshall.",
    "The Avalon mission revealed that Pantheon has infiltrated multiple levels of European intelligence.",
    "Givens was a key asset who provided the codes for the Lutetia vault before his disappearance.",
    "The 'Redlight' operation was a failed attempt to neutralize Jane Harrow in a Turkish safehouse.",
    "Marshall and Case utilized a prototype grappling hook during the infiltration of the Capitol Station.",
    "The Vorkuta connection suggests that older Soviet sleeper agents may be assisting Pantheon.",
    "Woods believes that the mole within the CIA has been active since the late 1980s."
]

# 2. User Question
user_query = str(input('Enter your response !!! '))

# 3. Math time: Turn text into numbers
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(sentences + [user_query])

# 4. Compare user query to our data
similarity = cosine_similarity(tfidf[-1], tfidf[:-1])
best_match_index = similarity.argsort()[0][-1]


best_score = similarity[0][best_match_index]

if best_score<0.2:
    print('CHATBOT : I cant find what you are looking for.')
else:
    print(f"CHATBOT : {sentences[best_match_index]}")
```
