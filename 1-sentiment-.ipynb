```python
from textblob import TextBlob

def analyze_sentiment(text):
    # Create a TextBlob object
    blob = TextBlob(text)
    
    # Get the polarity score (Polarity lies between [-1 to 1])
    polarity = blob.sentiment.polarity
    
    # Logic to categorize the score
    if polarity > 0:
        return "Positive ğŸ˜Š"
    elif polarity < 0:
        return "Negative ğŸ˜¡"
    else:
        return "Neutral ğŸ˜"   
```

```python
# Test it out!
user_input = input("Enter a sentence to analyze: ")
result = analyze_sentiment(user_input)

print(f"The sentiment is: {result}")
```
