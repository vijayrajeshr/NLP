```python
import re
# Use raw strings (r'') to avoid escape sequence warnings
p = re.compile(r'\d')
print(p.findall("I visited to college at 9 A.M. on 10th June 2022"))

p = re.compile(r'\d+')
print(p.findall("I visited to college at 9 A.M. on 10th June 2022 "))
```

```python
import re
p = re.compile(r'\w')
print(p.findall("He said * in some_lang."))

p = re.compile(r'\w+')
print(p.findall("I went to him at 11 A.M., he \
said *** in some_language."))

p = re.compile(r'\W')
print(p.findall("he said *** in some_language."))
```

```python
import re
# Correcting the SyntaxWarning by using r'\W+'
from re import split
print(split(r'\W+', 'Words, words , Words'))
print(split(r'\W+', "Word's words words"))
print(split(r'\W+', 'On 12th Jan 2016, at 11:02 AM'))
print(split(r'\d+', 'On 12th Jan 2016, at 11:02 AM'))
```

#8

```python
import re
regex = r"([a-zA-Z]+) (\d+)"
match = re.search(regex, "I was born on June 24")
if match != None:
    print ("Match at index %s, %s" % (match.start(), match.end()))
    print ("Full match: %s" % (match.group(0)))
    print ("Month: %s" % (match.group(1)))
    print ("Day: %s" % (match.group(2)))
else:
    print ("The regex pattern does not match.")
```

```python
import re
regex = r'^The'
strings = ['The quick brown fox', 'The lazy dog', 'A quick brown fox']
for string in strings:
    if re.match(regex, string):
        print(f'Matched: {string}')
    else:
        print(f'Not matched: {string}')
```
