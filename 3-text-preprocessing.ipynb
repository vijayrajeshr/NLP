```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.decomposition import TruncatedSVD
```

Download necessary NLTK Data

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
raw_docs = [
    

''' The history of Tamil Nadu is a vast timeline of cultural continuity, beginning with deep prehistoric roots. Archaeological findings at Attirampakkam suggest that hominin populations existed in the region as far back as 1.5 million years ago, long before modern Homo sapiens arrived. By the Mesolithic and Neolithic periods, humans in the region transitioned from nomadic hunter-gatherers using microlithic tools to settled communities that practiced agriculture and animal husbandry.'''

'''The Iron Age in Tamil Nadu is particularly noted for its megalithic burial sites, such as those found in Adichanallur, which reveal a sophisticated society that used iron tools, practiced elaborate funerary rites involving urns, and potentially used an early form of writing.'''

'''As the region entered the early historic period, often referred to as the Sangam era, it was defined by the reign of the "Three Crowned Kings": the Cheras, Cholas, and Pandyas. These dynasties presided over a landscape of shifting borders and constant rivalry, yet they fostered a golden age of Tamil literature and maritime trade. Their influence was recorded in the edicts of the Mauryan Emperor Ashoka and the inscriptions of King Kharavela of Kalinga, indicating their sovereignty remained intact despite the rise of northern empires. '''

'''This era saw the development of a complex social hierarchy involving local chieftains known as Velirs and the coexistence of diverse religious traditions, including Vedic practices alongside Jainism and Buddhism.'''

'''The medieval period witnessed the dramatic rise of the Imperial Cholas starting in the 9th century. Under leaders like Rajaraja I and Rajendra I, the Cholas transformed into a dominant maritime power, controlling the southern peninsula and extending their naval reach across the Bay of Bengal to Southeast Asia and Sri Lanka. This dominance eventually gave way to the rise of the Pandyan Empire and later the Vijayanagara Empire, which acted as a bulwark against northern incursions.'''

'''Following the fragmentation of Vijayanagara, local Nayaka governors established independent kingdoms in Madurai and Thanjavur. '''

'''By the late 18th century, European colonial interests, particularly the British East India Company, consolidated power to form the Madras Presidency. After India gained independence in 1947, the state underwent linguistic reorganization in 1956, eventually being renamed Tamil Nadu in 1969 to reflect its unique linguistic and cultural identity.'''


]


def preprocess_text(text):
    #Noise Reduction -----> remove punctuation and numbers.
    text = re.sub(r'[^a-zA-Z\s]',"",text)    
    
    # Normalization
    text = text.lower()
    
    #Tokenization
    tokens = word_tokenize(text)
    
    #Stop-Word Removal
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    
    print('\nTokens          :  ',tokens)
    print('\nStop Words      :  ',stop_words)
    print('\nfiltered_tokens :  ',filtered_tokens)  
    
    return" ".join(filtered_tokens)
```

#2 Execute Cleaning

```python
cleaned_docs = [preprocess_text(doc) for doc in raw_docs]
```

#3 Feature Extraction (TF - IDF)

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(cleaned_docs)
```

#4 Dimensionality Reduction

```python
#We reduce the features to 2 componenets for easy visualization/analysis
svd = TruncatedSVD(n_components=2)
reduced_data = svd.fit_transform(tfidf_matrix)
reduced_data
```
